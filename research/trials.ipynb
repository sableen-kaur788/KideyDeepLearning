```python
# Ensure annotations
def get_predict(x:int, y:int) -> int:
    return x*y

get_predict(x=2, y=4)
    
```

```python
get_predict(x=4, y="3") # '3333' print 3 four times.
```

```python
from ensure import ensure_annotations
@ensure_annotations
def get_pred(x:int, y:int) -> int:
    return x*y
```

```python
get_pred(x=4, y=3)
```

```python
get_pred(x=4, y="3")
```
